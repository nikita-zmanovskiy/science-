```python
# Jupyter notebook for interactive demonstration
# This would be a complete notebook in practice

import sys
sys.path.append('..')

from src.hierarchical_aggregator import HierarchicalAggregator
from src.data_generator import FinancialDataGenerator
import matplotlib.pyplot as plt
import numpy as np

# Quick demo cell
agg = HierarchicalAggregator(base_delta=1.0, max_timeframe=300.0)
generator = FinancialDataGenerator()

# Process some data
prices = []
timestamps = []

for i in range(10000):
    timestamp = i * 0.1  # 10 Hz data
    price = generator._generate_value(timestamp)
    agg.update(timestamp, price)
    prices.append(price)
    timestamps.append(timestamp)

print(f"Processed {len(prices)} events")
print(f"Memory usage: {agg.memory_usage} blocks")
```
